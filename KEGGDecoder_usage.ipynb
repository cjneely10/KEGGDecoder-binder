<h1>KEGG decoder examples and usage</h1>

    Designed to parse through a KEGG-Koala outputs (including blastKOALA, ghostKOALA, KOFAMSCAN) to determine the completeness of various metabolic pathways. KEGGDecoder was orginally designed by Ben Tull and is available at: https://github.com/bjtully/BioData/tree/master/KEGGDecoder.
    This package is designed to easily visualize genome/MAGs metabolic capabilities through the outputs of KEGGDecoder.
    
    



    It is based off of scripts and work done by Ben Tully, available at: https://github.com/bjtully/BioData/tree/master/KEGGDecoder and modified by Ben Tully, Roth Conrad, Taylor Reitner, Luiz Irber, and Jay Osvatic.
   

<strong>Possible outputs</strong>

4 different outputs can be produced by this repository. By using the option --vizoption and 'static/interactive/tanglegram' you can pick the desired output.
    
    1.The FUNCTION_OUT.list generates a TSV version of the heat map. The first row contains pathway/process
    names, subsequent rows contain submitted groups/genomes and fractional percentage of pathway/process

    2. A 'static' figure output is an SVG file function_heatmap.svg. Each distinct identifier before the
    underscore in the FASTA file will have a row

    3. A 'interactive' figure output is an HTML file function_heatmap.html. Each distinct identifier before 
    the underscore in the FASTA file will have a row, but can be loaded into a browser and value will be displayed
    by hovering over a cell with the mouse. Draw a box to zoom in on specific regions.

    4.'tanglegram' -- under construction


<strong>Input file generation and formatting for functional heatmap </strong>

The functional heatmaps are generated by the default output of BlastKOALA, GhostKOALA or KOFAMscan.

Start with protein FASTA file (INPUT_PROTEIN.fasta). This file can be multiple genomes combined. Be sure your submitted FASTA file has headers that group genomes together, KEGG-decoder.py groups based on the name provided in FASTA header before the first underscore (_)

Process protein sequences through KEGG-KOALA (GhostKoala, BlastKoala, or KOFAMSCAN) and download the tab-delimited KO assignment text file (KOALA_OUTPUT.txt)
